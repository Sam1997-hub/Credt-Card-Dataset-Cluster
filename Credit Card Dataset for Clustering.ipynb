```python
from sklearn.impute import SimpleImputer
```

```python
si=SimpleImputer(strategy='median')
```

```python
df[['CREDIT_LIMIT','MINIMUM_PAYMENTS']]=si.fit_transform(df[['CREDIT_LIMIT','MINIMUM_PAYMENTS']])
```

```python
df.describe(include='all').T
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(round(df.corr(),2),annot=True,fmt='0.5g')
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA()
```

```python
pca.fit(df)
```

```python
pca.explained_variance_ratio_
```

```python
plt.plot(range(1,18),pca.explained_variance_ratio_.cumsum(),marker='o',linestyle='--')
plt.grid()
plt.show()
```

```python
pca=PCA(n_components=3)
```

```python
df_pca=pca.fit_transform(df)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans()
from sklearn.metrics import silhouette_score
```

```python


for i in range(1,25):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df_pca)
    print(i,"cluster:",silhouette_score(df_pca,kmeans.labels_))
```

```python
wcss=[]
for i in range(1,25):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df_pca)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,25),wcss,marker='o',linestyle='--')
plt.grid()
plt.show()
```

```python
## number of cluster =2,Based upon Silhouette score and Weighted Cluster Sum of Squares Graph
```

```python
kmeans=KMeans(n_clusters=2)
```

```python
kmeans.fit(df_pca)
```

```python
df['Labels']=kmeans.labels_
```

```python
df.groupby(['Labels']).mean()
```

```python
df['Cluster']=np.where(df['Labels']==1,"High Spender","Low Spender")
```

```python
df.groupby(['Cluster']).mean()
```

```python
sns.scatterplot(df['CREDIT_LIMIT'],df['PURCHASES'],hue =df['Cluster'])
```
